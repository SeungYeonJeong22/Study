```python
import pandas as pd
import matplotlib.pyplot as plt

chicken07 = pd.read_csv('./data/chicken_07.csv')
chicken08 = pd.read_csv('./data/chicken_08.csv')
chicken09 = pd.read_csv('./data/chicken_09.csv')
chicken_data = pd.concat([chicken07,chicken08,chicken09])
print(type(chicken_data))

chicken_data.reset_index(drop = True)
weeks = ['월','화','수','목','금','토','일']
ages = list(set(chicken_data['연령대']))

plt.rcParams['font.family'] = 'Malgun Gothic'

result_data = chicken_data.groupby('연령대').sum()['통화건수'].sort_values(ascending = False)
# result_data = sum_data.sort_values(ascending = False)

plt.figure(figsize = (7,5))
plt.bar(result_data.index,result_data.values)
plt.title('연령대별 치킨 주문량')
plt.show()
```

```python
result_data = chicken_data.groupby('시군구').sum()['통화건수'].sort_values(ascending=False)
plt.figure(figsize = (40,10))
plt.bar(result_data.index,result_data.values)
plt.title('시군구별 치킨 주문량')
plt.show()

```

```python


result_data = chicken_data.groupby('시군구').sum()['통화건수'].sort_values(ascending=False)
n = list(result_data.values)

plt.figure(figsize = (40,10))
plt.bar(result_data.index,result_data.values)
plt.xticks(rotation = 30)
plt.title('시군구별 치킨 주문량')

for i, txt in enumerate(n):
    plt.annotate(txt,(result_data.index[i],result_data.values[i]))
    

plt.show()
```
