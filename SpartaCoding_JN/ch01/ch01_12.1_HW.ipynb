```python
import pandas as pd
import matplotlib.pyplot as plt

pizza_data = pd.read_csv('./data/pizza_09.csv')
print(pizza_data.columns)
```

```python
week_data = pizza_data.groupby('요일')['통화건수']
print(week_data.head(1))
```

```python
week_data = pizza_data.groupby('요일').sum()['통화건수'].sort_values(ascending = True)
print(week_data)
```

```python
plt.rcParams['font.family'] = 'Malgun gothic'
n = list(week_data.values)
plt.figure(figsize = (8,5))
plt.bar(week_data.index,week_data.values)
plt.title('요일 따른 피자 주문량 합계')

for i,txt in enumerate(n):
    plt.annotate(txt,(week_data.index[i],week_data.values[i]))

plt.show()
```

```python
chicken07 = pd.read_csv('./data/chicken_07.csv')
chicken08 = pd.read_csv('./data/chicken_08.csv')
chicken09 = pd.read_csv('./data/chicken_09.csv')

chicken_data = pd.concat([chicken07,chicken08,chicken09])

# weeks = ['월','화','수','목','금','토','일']
# weekC_data = chicken09.groupby('요일').sum()['통화건수'].reindex(weeks)
# weekP_data = pizza_data.groupby('요일').sum()['통화건수'].reindex(weeks)

weekC_data = chicken09.groupby('요일').sum()['통화건수'].sort_values()
weekP_data = pizza_data.groupby('요일').sum()['통화건수'].sort_values(ascending = False)

c = list(weekC_data.values)
p = list(weekP_data.values)

for i,txt in enumerate(c):
    plt.annotate(txt,(weekC_data.index[i],weekC_data.values[i]))
    
for i,txt in enumerate(p):
    plt.annotate(txt,(weekP_data.index[i],weekP_data.values[i]))


plt.bar(weekC_data.index,weekC_data.values)
plt.bar(weekP_data.index,weekP_data.values,color = 'red')
plt.title('요일에 따른 치킨과 피자 주문량 합계')

plt.show()
```
